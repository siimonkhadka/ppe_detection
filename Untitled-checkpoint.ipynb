```python
import cv2
import os
import random
from ultralytics import YOLO
```

```python
model = YOLO('C:/Users/user/minor_project/runs/train/my_yolov8_model3/weights/best.pt')

# Class-based color mapping
COLOR_MAP = {
    'Helmet': (0, 255, 0),      # Green
    'Vest': (0, 255, 0),        # Green
    'NoHelmet': (0, 0, 255),    # Red
    'NoVest': (0, 0, 255),      # Red
    'Person': (255, 255, 0)     # Yellow 
}

image_dirs = [
    'C:/Users/user/minor_project/const_images',
]

```

```python
image_paths = []
for folder in image_dirs:
    image_paths += [os.path.join(folder, f) for f in os.listdir(folder) if f.lower().endswith(('.jpg', '.png'))]

if not image_paths:
    raise FileNotFoundError("No images found in given folders.")

# Pick a random image for inference (or loop for all images if you want)
image_path = random.choice(image_paths)
print(f"Using image: {image_path}")

# Run inference with lower confidence threshold (adjust if needed)
img = cv2.imread(image_path)
results = model(img, conf=0.1)  # Lowered confidence to 0.1 for better detection
pred_img = img.copy()

occupied_regions = []

for box in results[0].boxes:
    cls_id = int(box.cls[0])
    cls_name = model.names[cls_id]
    conf = float(box.conf[0])

    # Debug print to see what classes are detected and their confidence
    print(f"Detected: {cls_name} with confidence {conf:.2f}")

    # Skip drawing for 'Person' class
    if cls_name == 'Person':
        continue

    x1, y1, x2, y2 = map(int, box.xyxy[0])

    color = COLOR_MAP.get(cls_name, (255, 255, 255))  # Default white (shouldn't happen)
    label = f"{cls_name} {conf:.2f}"

    # Draw bounding box for vest, helmet, nohelmet, novest only
    cv2.rectangle(pred_img, (x1, y1), (x2, y2), color, 2)

    # Calculate text size
    text_size, _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 2)
    text_w, text_h = text_size

    y_text = y1 - 10 if y1 - 10 > text_h else y1 + text_h + 10
    x_text = max(0, min(x1, pred_img.shape[1] - text_w))

    # Adjust text position to avoid overlap
    for (ox1, oy1, ox2, oy2) in occupied_regions:
        if (x_text < ox2 and x_text + text_w > ox1) and (y_text < oy2 and y_text + text_h > oy1):
            y_text = oy2 + 5

    occupied_regions.append((x_text, y_text, x_text + text_w, y_text + text_h))

    # Put label text
    cv2.putText(pred_img, label, (x_text, y_text), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

# Save output
os.makedirs('inference_output', exist_ok=True)
output_path = os.path.join('inference_output', os.path.basename(image_path))
cv2.imwrite(output_path, pred_img)

print(f"Output saved to: {output_path}")
```

```python
image_dirs = ['C:/Users/user/minor_project/const_images']  
output_dir = 'inference_output'
os.makedirs(output_dir, exist_ok=True)

image_paths = []
for folder in image_dirs:
    image_paths += [
        os.path.join(folder, f) for f in os.listdir(folder) 
        if f.lower().endswith(('.jpg', '.png','.jpeg'))
    ]

print(f"Total images found: {len(image_paths)}\n")

# ****Inference loop ***
for idx, image_path in enumerate(image_paths):
    print(f"[{idx + 1}] Processing: {image_path}")

    img = cv2.imread(image_path)
    if img is None:
        print(f" Could not read image: {image_path}")
        continue

    results = model(img, conf=0.1)  # Lower confidence for missed detections
    pred_img = img.copy()

    for box in results[0].boxes:
        cls_name = model.names[int(box.cls[0])]
        if cls_name == 'Person':
            continue  # Skip bounding box for person

        conf = float(box.conf[0])
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        label = f"{cls_name} {conf:.2f}"
        color = COLOR_MAP.get(cls_name, (255, 255, 255))  # Fallback: white

        cv2.rectangle(pred_img, (x1, y1), (x2, y2), color, 2)
        cv2.putText(pred_img, label, (x1, max(20, y1 - 10)), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Save result image
    output_path = os.path.join(output_dir, os.path.basename(image_path))
    success = cv2.imwrite(output_path, pred_img)
    if success:
        print(f"Saved: {output_path}")
    else:
        print(f" Failed to save: {output_path}")

print("\n Folder inference complete.")
```

```python
import os
import cv2
from ultralytics import YOLO

model = YOLO('C:/Users/user/minor_project/runs/train/my_yolov8_model3/weights/best.pt')
image_dirs = ['C:/Users/user/minor_project/const_images']

# Your model must be loaded already as `model`

# Collect image paths again
image_paths = []
for folder in image_dirs:
    image_paths += [
        os.path.join(folder, f) for f in os.listdir(folder)
        if f.lower().endswith(('.jpg', '.jpeg', '.png'))
    ]

print(f"Total images to analyze: {len(image_paths)}\n")

class_counts = {}
class_confidences = {}
total_detections = 0

for idx, image_path in enumerate(image_paths):
    print(f"[{idx + 1}] Processing: {image_path}")

    img = cv2.imread(image_path)
    if img is None:
        print(f" Could not read image: {image_path}")
        continue

    results = model(img, conf=0.5)

    for box in results[0].boxes:
        cls_name = model.names[int(box.cls[0])]
        if cls_name == 'Person':  # skip person class
            continue

        conf = float(box.conf[0])
        class_counts[cls_name] = class_counts.get(cls_name, 0) + 1
        class_confidences[cls_name] = class_confidences.get(cls_name, 0) + conf
        total_detections += 1

print("\n===== Inference Summary =====")
print(f"Total images processed: {len(image_paths)}")
print(f"Total PPE detections (excluding 'Person'): {total_detections}")

print("\nDetections per class with average confidence:")
for cls_name in class_counts:
    avg_conf = class_confidences[cls_name] / class_counts[cls_name]
    print(f"  - {cls_name}: {class_counts[cls_name]} detections, average confidence {avg_conf:.2f}")
```
